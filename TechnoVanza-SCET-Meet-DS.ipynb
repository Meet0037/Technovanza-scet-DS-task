# **Question 1 : Which are the movies with the third lowest and the third highest budget?**

```python
def find_third_low_high(x):
    #use the function 'idxmin' to find the index of lowest budget movie.
    min_index = data[x].idxmin() * 3
    #use the function 'idxmax' to find the index of Highest highest movie.
    high_index = data[x].idxmax() * 3
    high = pd.DataFrame(data.loc[high_index,:]) 
    low = pd.DataFrame(data.loc[min_index,:])
    
    #print the movie with third high and third low budget
    print("Movie Which Has Third Highest "+ x + " : ",data['original_title'][high_index])
    print("Movie Which Has Third Lowest "+ x + "  : ",data['original_title'][min_index])
    return pd.concat([high,low],axis = 1)

data['budget'] = data['budget'].replace(0,np.NAN)
find_third_low_high('budget')
```

We can derive top 10 highest budget movies.

```python
##sort the 'budget' column in decending order and store it in the new dataframe.
info = pd.DataFrame(data['budget'].sort_values(ascending = False))
info['original_title'] = data['original_title']
data1 = list(map(str,(info['original_title'])))

#extract the top 10 budget movies data from the list and dataframe.
x = list(data1[:10])
y = list(info['budget'][:10])

#plot the figure and setup the title and labels.
ax = sns.pointplot(x=y,y=x,color='g')
sns.set(rc={'figure.figsize':(10,5)})
ax.set_title("Top 10 High Budget Movies",fontsize = 15)
ax.set_xlabel("Budget",fontsize = 13)
sns.set_style("darkgrid")
```

# **Question 2. Which are the movies with the most and the least earned revenue?**

```python
def find_most_least_revenue(x):
    #use the function 'idxmin' to find the index of lowest revenue movie.
    min_index = data[x].idxmin() 
    #use the function 'idxmax' to find the index of Highest revenue movie.
    high_index = data[x].idxmax()
    high = pd.DataFrame(data.loc[high_index,:]) 
    low = pd.DataFrame(data.loc[min_index,:])
    
    #print the movie with third high and third low revenue
    print("Movie Which Has Highest "+ x + " : ",data['original_title'][high_index])
    print("Movie Which Has Lowest "+ x + "  : ",data['original_title'][min_index])
    return pd.concat([high,low],axis = 1)

find_most_least_revenue('revenue')
```

We can derive top 10 highest revenue movies.

```python
#sort the 'revenue' column in decending order and store it in the new dataframe.
info = pd.DataFrame(data['revenue'].sort_values(ascending = False))
info['original_title'] = data['original_title']
data2 = list(map(str,(info['original_title'])))

#extract the top 10 movies with high revenue data from the list and dataframe.
x = list(data2[:10])
y = list(info['revenue'][:10])

#make the point plot and setup the title and labels.
ax = sns.pointplot(x=y,y=x,color='y')
sns.set(rc={'figure.figsize':(10,5)})
ax.set_title("Top 10 High Revenue Movies",fontsize = 15)
ax.set_xlabel("Revenue",fontsize = 13)
sns.set_style("darkgrid")
```

# **Question 3. What is the average runtime of movies in the year 2006?**

```python
data['release_year']=pd.to_datetime(data['release_year'])
avgruntime = data.groupby(np.floor((data['release_year'].dt.year-2006)/10)).mean()['runtime']
print("Average runtime of movies in the year 2006",avgruntime)
```

So our runtime of movies in the year 2006 is **102.07179**

*We can also derive,*

```python
#extract the month number from the release date.
month_release = data['release_date'].dt.month

#count the movies in each month using value_counts().
number_of_release = month_release.value_counts().sort_index()
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
number_of_release = pd.DataFrame(number_of_release)
number_of_release['month'] = months

#change the column name of the new dataframe 'number_of_release'
number_of_release.rename(columns = {'release_date':'number_of_release'},inplace=True)

#plot the bar graph using plot.
number_of_release.plot(x='month',kind='bar',fontsize = 11,figsize=(8,6),color='purple')

#set the labels and titles of the plot.
plt.title('Months vs Number Of Movie Releases',fontsize = 15)
plt.xlabel('Month',fontsize = 13)
plt.ylabel('Number of movie releases',fontsize = 13)
sns.set_style("darkgrid")
```

According to the plot we can conclude that there are higher number of release in september and october month.

```python
month_release = pd.DataFrame(month_release)

#change the column name of the new dataframe 'month_release'.
month_release.rename(columns = {'release_date':'release_month'},inplace=True)

#add a new column 'revenue' in the dataframe 'month_release'.
month_release['revenue'] = data['revenue']

#make the group of the data according to their month and calculate the mean revenue of each month.
mean_revenue  = month_release.groupby('release_month').mean()
mean_revenue['month'] = months

#make the bar plot using pandas plot function.
mean_revenue.plot(x='month',kind='bar',figsize = (8,6),fontsize=11,color='pink')

#setup the title and lables of the plot.
plt.title('Average revenue by month (1960 - 2015)',fontsize = 15)
plt.xlabel('Month',fontsize = 13)
plt.ylabel('Average Revenue',fontsize = 13)
sns.set_style("darkgrid")
```

According to the plot we can say that movie which release in may or june month made the high revenue in comparison of other month release.

If we talk about Genre then,

```python
def count_genre(x):
    #concatenate all the rows of the genrs.
    data_plot = data[x].str.cat(sep = '|')
    data3 = pd.Series(data_plot.split('|'))
    #conts each of the genre and return.
    info = data3.value_counts(ascending=False)
    return info

#call the function for counting the movies of each genre.
total_genre_movies = count_genre('genres')
#plot a 'barh' plot using plot function for 'genre vs number of movies'.
total_genre_movies.plot(kind= 'barh',figsize = (13,6),fontsize=12,colormap='Wistia_r')

#setup the title and the labels of the plot.
plt.title("Genre With Highest Release",fontsize=15)
plt.xlabel('Number Of Movies',fontsize=13)
plt.ylabel("Genres",fontsize= 13)
sns.set_style("whitegrid")
```

*We can also represent* **director who directs maximum movies**

```python
count_director_movies = count_genre('director')

#plot a bar graph
count_director_movies.iloc[:20].plot(kind='bar',figsize=(13,6),fontsize=12,color="lightgreen")

#setup the title and the labels 
plt.title("Director Vs Number Of Movies",fontsize=15)
plt.xticks(rotation=70)
plt.ylabel("Number Of Movies",fontsize= 13)
sns.set_style("ticks")
```
